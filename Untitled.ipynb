```python
# ex_bills = {}
# inc_bills = {}

class Calculator:
    def __init__(self,ex_bills ={}, inc_bills ={}):
        self.ex_bills = ex_bills
        self.inc_bills = inc_bills
        self.add_income()
        self.add_propex()
        self.total_inc()
        self.total_expenses()
        self.calc_cash_flow()
        self.summary()
        print(self.calc_roi())
        
    
    def add_income(self):
        while True:
            ask_income = input ("Do you have any income you would like to add? (yes/no/ see types of income)").lower().strip()
            if ask_income == 'see':
                print("=" *70)
                print("Types of income you can include are: \nRental Income\nLaundrt\nStorage\nMisc")
            if ask_income == 'yes':
                income_name = input("What would you like to add?  ")
                income_cost = int(input("How much is this expense a month? "))
                self.inc_bills[income_name] = income_cost
            elif ask_income == 'no':
               # print("You have no income to report.")
                break
            else:
                print("Please enter a valid response.")
                
    def total_inc(self):
        inc_values = self.inc_bills.values()
        inc_total = sum (inc_values)
        return inc_total
    
    def add_propex(self):
        while True:
            ask_ex = input ("Do you have any expenses you would like to add? (yes/no/ see types of expenses)").lower().strip()
            if ask_ex == 'see':
                print("=" *70)
                print("Types of expenses you can include are: \nTaxes\nInsurance\nHOA\nSnow\nLawn\nVacancy\nRepairs\nCapEx\nProperty Management\nMortgage")
            if ask_ex == 'yes':
                ex_name = input("What would you like to add?  ")
                ex_cost = int(input("How much is this expense a month? "))
                self.ex_bills[ex_name] = ex_cost
            elif ask_ex == 'no':
                #print("You have no expenses to report.")
                break
            else:
                print("Please enter a valid response.")
                    
    def total_expenses(self):
        values = self.ex_bills.values()
        ex_total = sum (values)
        return ex_total
    
    def calc_cash_flow(self):
        total_cash_flow = self.total_inc() - self.total_expenses()
        return total_cash_flow
    
    def total_investment(self):
        down_payment = float(input("How much did you put for a down payment? "))
        closing_cost = float(input("How much were your closing costs? "))
        repair_money = float(input("Do you have any money for repairs? "))
        total_investment = down_payment + closing_cost + repair_money
        return total_investment 
                
        
    def calc_roi(self):
        monthly_cash = self.calc_cash_flow() *12
        roi = monthly_cash/ self.total_investment()
        return roi
    
    def summary(self):
        print(f'Your incomes are as follows:')
        for k, v in self.inc_bills.items():
            print(f'{k}')
            print(f'{v}')
        print(f'Your expenses are as follows:')
        for k, v in self.ex_bills.items():
            print(f'{k}')
            print(f'{v}')
        print(f'The percentage of your return investment is {self.calc_roi()}')
        
        
        
```
